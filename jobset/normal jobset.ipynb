```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
import random

seed_value=3
random.seed(seed_value)

distribution="normal" # 

#
j_mean = 5  # 
j_std_dev = 1  # 
j_size = 1000  # 

p_mean = 5  # 
p_std_dev = 1  # 
p_size = 1000  # 

m_low=1 #
m_high=5 #

t_mean = 15  # 
t_std_dev = 3  # 
t_size = 1000  # 

gene=10000
for g in np.arange(0,gene):
    
    #
    j_data0 = np.random.normal(j_mean, j_std_dev, j_size)
    j_data = np.round(j_data0).astype(int)
    j_number=j_data[np.random.randint(0,1000)] # 
    #print("",j_number)

    #
    p_data0 = np.random.normal(p_mean, p_std_dev, p_size)
    p_data = np.round(p_data0).astype(int)
    p_number=p_data[:j_number]#
    #print("",p_number)
    p_number_max=max(p_number)
    #print("",p_number_max)

    #
    MS_J=np.zeros((int(j_number),int(p_number_max)))
    for i in np.arange(0,j_number):
        k=int(p_number[i])
        for j in np.arange(0,k):
            MS_J[i,j]=np.random.randint(m_low,m_high,1)
    #print(':\n',MS_J)

    #
    T_J=np.zeros((int(j_number),int(p_number_max)))
    t_data0 = np.random.normal(t_mean, t_std_dev, t_size)
    t_data = np.round(t_data0).astype(int)

    for i in np.arange(0,j_number):
        k=int(p_number[i])
        for j in np.arange(0,k):
            T_J[i,j]=t_data[np.random.randint(0,1000)]
    #print(':\n',T_J)

    JOBSET_0=np.zeros((int(j_number),int(p_number_max)))
    JOBSET=pd.DataFrame(JOBSET_0, columns=[f"process{i+1}" for i in range(int(p_number_max))], index=[f"job{i+1}" for i in range(int(j_number))])
    JOBSET=JOBSET.astype(str)

    for i in np.arange(0,j_number):
        k=int(p_number[i])
        for j in np.arange(0,k):
            m=MS_J[i,j]
            t=T_J[i,j]
            JOBSET.iloc[i,j]=str(m),str(t)    
    #print('\n',JOBSET)
    
    str1="D://pysrc/wang_data/jobset/"+str(distribution)+"/"
    j_element="j_mean["+str(j_mean)+"],j_std_dev["+str(j_std_dev)+"]/"
    p_element="p_mean["+str(p_mean)+"],p_std_dev["+str(p_std_dev)+"]/"
    m_element="machine["+str(m_high-1)+"]/"
    t_element="t_mean["+str(t_mean)+"],t_std_dev["+str(t_std_dev)+"]/"
    file="seed["+str(seed_value)+"]/"
    name="["+str(j_number)+']r['+str(p_number_max)+']c,'+str(g)+'gene'
    str2=".csv"

    filename=str1+j_element+p_element+m_element+t_element+file+name+str2
    
    JOBSET.to_csv(filename,index=True)
```
