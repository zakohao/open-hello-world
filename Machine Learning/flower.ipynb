```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_csv('D://vscode//data//iris.csv')
display(data)
```

```python
#とりあえず標準化前に散布図を描いてみる
import matplotlib.pyplot as plt
# Create a scatter plot
fig, ax = plt.subplots()
ax.scatter(data['petal_length'], data['petal_width'])
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
#標準化したい列名を設定
scaling_columns = ['petal_length','petal_width'] 
#列ごとに標準化を行う
sc = StandardScaler().fit(data[scaling_columns])

#sc = StandardScaler().fit(data[scaling_columns])
# 標準化したカラムのみ元のDataFrameに戻す
scaled_train = pd.DataFrame(sc.transform(data[scaling_columns]), columns=scaling_columns, index=data.index)
data.update(scaled_train)
data
```
