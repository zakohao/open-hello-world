```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
import random

random.seed(5)

j_low=3 #ジョブ数の最小
j_high=10 #ジョブ数最大
p_low=3 #プロセス数の最小
p_high=8 #プロセス数最大
m_low=1 #最小機械番号
m_high=5 #最大機械番号
t_low=10 #最小プロセス時間
t_high=20 #最大プロセス時間

j_number=np.random.randint(j_low,j_high,1)#ジョブセットの中のジョブ数を固定
print('ジョブセットのジョブ数:',j_number)
p_number=np.zeros(j_number)#各ジョブのプロセス数を固定
#print('p_number:',p_number)
for i in np.arange(0,j_number):
    p_number[i]=np.random.randint(p_low,p_high,1)
print('各ジョブのプロセス数:',p_number)
p_number_max=max(p_number)
print('各ジョブの最大プロセス:',p_number_max)

MS_J=np.zeros((int(j_number),int(p_number_max)))
for i in np.arange(0,j_number):
    k=int(p_number[i])
    #print('今は',i+1,'行目，全部',k,'個プロセス')
    for j in np.arange(0,k):
        MS_J[i,j]=np.random.randint(m_low,m_high,1)
print('ジョブセットの機械番号:\n',MS_J)

T_J=np.zeros((int(j_number),int(p_number_max)))
for i in np.arange(0,j_number):
    k=int(p_number[i])
    #print('今は',i+1,'行目，全部',k,'個プロセス')
    for j in np.arange(0,k):
        T_J[i,j]=np.random.randint(t_low,t_high,1)
print('ジョブセットの操作時間:\n',T_J)

JOBSET_0=np.zeros((int(j_number),int(p_number_max)))
JOBSET=pd.DataFrame(JOBSET_0, columns=[f"process{i+1}" for i in range(int(p_number_max))], index=[f"job{i+1}" for i in range(int(j_number))])
JOBSET=JOBSET.astype(str)
#print(JOBSET)
for i in np.arange(0,j_number):
    k=int(p_number[i])
    for j in np.arange(0,k):
        m=MS_J[i,j]
        t=T_J[i,j]
        JOBSET.iloc[i,j]=str(m),str(t)
        
print('自動生成したジョブセット：\n',JOBSET)

#np.savetxt(fname='d:/pysrc/wang_data/automatic generate jobset.csv',X=JOBSET,fmt='%0d,%0d,%0d,%0d,%0d,%0d,%0d',delimiter=',')
JOBSET.to_csv('d:/pysrc/wang_data/automatic generate jobset.csv',index=True)
```
